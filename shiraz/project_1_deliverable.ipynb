## **The Chordonomicon Dataset: Data Description and Provenance**

The Chordonomicon is a large-scale dataset designed for deep learning applications in music, containing over **666,000 songs** and their chord progressions. It was created to address the lack of large, publicly available datasets for chord progressions, which remain an underexplored data domain despite their importance in music composition.

### **Data Description**

The dataset primarily features the symbolic representation of music compositions through **chord progressions**. Each chord progression consists of an average of **76.48 chords**, with a median of 71. The total dataset contains 679,807 unique musical tracks and $51,994,634$ total chords.

The data is annotated with:

* **Structural Parts** (Semantic Labels): Eight distinct part categories are used: **Intro**, **Verse**, **Chorus**, **Bridge**, **Interlude**, **Solo**, **Instrumental**, and **Outro**. The dataset provides $2,670,457$ **structural labels** for $397,580$ music tracks.

    * **Distribution:** **Verse** and **Chorus** are the most frequent parts.

* **Metadata:**

    * **Genre** (and Sub-genre): **12 main genres** were categorized from more specific Spotify genres to characterize the tracks' main genre. There are $352,111$ tracks with a main genre and $145,218$ tracks with a rock sub-genre.

    * **Release Date:** Encoded by **decade** for a multi-class classification problem with 15 different decades, with the majority of tracks from 2010. $422,181$ tracks have a release date.

    * **Spotify IDs:** IDs are provided for tracks and artists.

* **Representations:**

    * **Symbolic Chord Progression:** Chords are converted into the **Harte syntax** and the syntax proposed by the **Functional Harmony Ontology (FHO)** for interlinkability with existing knowledge frameworks.

    * **Graph Representation:** Each music track is transformed into a **weighted directed graph** where nodes are chords and edges are chord progressions, with weights reflecting the progression's frequency. The dataset includes $667,858$ graphs.

    * **Chord Statistics:** The most frequent chords are **natural triads**, particularly the "G" chord, which is common in modern, user-generated music. The chord types are predominantly **Major** ($37,031,150$ occurrences) and **Minor** ($13,368,240$ occurrences). Most chords consist of three notes ($46,529,077$ occurrences).

### **Missing Data and Analysis Impact**

The dataset has missing annotations for some tracks, which could impact analyses focused on those features:

* **Missing Structural Parts:** Only $397,580$ tracks out of $679,807$ have structural part annotations, meaning $282,227$ tracks (about 41.5%) have only the chord progression retained.

* **Missing Metadata:**

    * $257,626$ tracks (about 37.9%) are missing a **release date**.

    * $327,696$ tracks (about 48.2%) are missing a **main genre**.

    * Tracks that did not fall into one of the 12 most common genres were **excluded from the genre classification task**.
    
These gaps mainly affect tasks relying on complete metadata and structural analysis, such as the decade or genre classification, as well as any task that specifically requires the internal song structure (e.g., sequence modeling that is aware of the structural part boundaries).

### **Provenance: Data Collection and Purpose**

The Chordonomicon dataset was collected and presented by **Spyridon Kantarelis**, **Konstantinos Thomas**, **Vassilis Lyberatos**, **Edmund Dervakos**, and **Giorgos Stamou** from the Artificial Intelligence and Learning Systems Laboratory, National Technical University of Athens, Greece.



The data was collected by **scraping various sources of user-generated progressions and associated metadata**. Specifically, the researchers leveraged web scraping techniques to extract song chord data from the **Ultimate Guitar (UG) platform**. They adhered to UG's `robots.txt` and Terms of Service, obtaining an initial $1.4$ million song chord pages and consolidating them into $793,362$ unique songs. The "best" version of a song was retained, determined by a weighted average of user rating and vote count.

The primary **reason for collecting the data** was:

* **To create the largest chord progression dataset to date** and fill the lack of large-scale datasets suitable for deep learning applications in music.

* **To provide a resource** for exploring advanced machine learning techniques (like transformers and graph machine learning) and hybrid systems that combine knowledge representation and machine learning.

* **To facilitate further advancements** in chord progression research, particularly in the domain of chord recognition, by providing more than twenty times the amount of chord progression data than the largest prior dataset.

* **To provide valuable insights** to the research community and serve as an ideal testbed due to the unique ability of chord progressions to be represented in multiple formats and the wealth of information they convey.

### **Phenomenon Modeled: Music Chord Progressions**

The phenomenon being modeled is **music chord progressions**. Chord progressions are a sequence of musical chords that typically form the harmonic **backbone of a musical composition**. They encapsulate important information about the music, relating to both its **structure and conveyed emotions**. In many cases, a chord progression is the sole information a musician needs to play along and follow a piece of music.

##### **Background and Definitions**

* **Chords:** A music chord is generally defined as a **group of notes played together**. In Western music, chords are commonly built on the interval of **thirds (tertian chords)**, and their distinct types arise from the quality of their intervals.

    * **Triads and Extended Chords:** Chords typically comprise **three notes (triads)**, while those with four to seven notes are termed "extended" chords. The "power" chord, common in metal music, is an exception, featuring only **two notes** with a fifth interval.

    * **Chord Progressions:** This is the resulting sequence of chords. They convey wealth of information, such as their harmonic function.

#### **Main Features and Data Support**

The Chordonomicon dataset models this phenomenon by representing **679,807 unique musical tracks** with their symbolic chord progressions. The main features of the modeled chord progressions are:

* **Length:** The chord progressions are quite long, consisting of an average of **76.48 chords**, with a median of 71 chords.

* **Dominant Chords:** The dataset comprises $51,994,634$ **total chords**. **Natural triads**, particularly the **"G" chord**, are the most frequently occurring, followed by "C" and "D".

    * This prevalence is thought to be due to the simplicity favored in modern, user-generated music and the widespread use of capos.

* **Chord Types and Notes:** The most common chord types are **Major** ($37,031,150$ occurrences) and **Minor** ($13,368,240$ occurrences). Most chords in the dataset are **three-note chords** ($46,529,077$ occurrences).

* **Structure:** $397,580$ tracks have structural part annotations (a total of $2,670,457$ parts). The eight distinct part categories used are: **Intro**, **Verse**, **Chorus**, **Bridge**, **Interlude**, **Solo**, **Instrumental**, and **Outro**. The **Verse** and **Chorus** are the most frequent parts. 

#### **Analysis Model Detail: Markov Transition Model**

Our analysis is modeling the **Markov Transition** of the chord sequence, with a specific focus on **chorus progression**.

* **Model Goal:** To find the **transition probabilities** between chords, including the probability of stopping (ending the chorus).

* **Modeling Limitation:** The progression is artificially cut off at **16 chords** to limit the length, as few choruses exceed that length. This cut-off length is a **hyperparameter**.

* **Data Curation for Modeling:**

    1. **Bass Note Removal:** The letter after the '/' (the bass note) was removed because its absence **"doesn't really change the chord's sound that much,"** simplifying the model from one based on the full chord to only the non-bass portion.

    2. **Chord Replacement Map:** To handle the excessive number of distinct chords, a replacement map was created to substitute the **least frequent chords** with more robust equivalents. This involved replacing them with chords that **sound extremely similar** or have a **more common name**, solving the problem of too many distinct states for a robust model.
